```python
import pandas as pd
import numpy as np
import os
import glob
import re
import math
cwd=os.getcwd()
print("Current working directory",cwd)
filenames = glob.glob(cwd + "/*.csv")
main_data=[]

for datapath in filenames:
    print("File Readind Started..........")
    print(filenames)
    dataset=pd.read_csv(datapath)
    ### ranking dictionary
    s=dataset.shape[0]
    ls2=[]
    if s/24==int(s/24):
        n=int(s/24)
        for x in range(n):
            for y in range(1,25,1):
                ls2.append(y)
               
    elif s<24:
        for x in range(1,s+1,1):
            ls2.append(x)
            
    else:
        for x in range(int(s/24)):
            for y in range(1,25,1):
                ls2.append(y)
                
        for x in range(s-int(s/24)*24):
            ls2.append(x+1)
    dataset["Rank"]=[x for x in ls2]
    dataset.rename({"_root":"root","Data Crawl Completion Time(UTC)":"date",
                    "Crawl Source Input":"Source",
                   "Brand Input":"Brand"},axis=1,inplace=True)
    dataset["Source"]="Amazon"
    ls1=["Redmi","Samsung","Poco","Realme","Oneplus","Narzo","Infinix","Iqoo","Tecno","Xiaomi","Oppo","Vivo"]
    brand=[x.lower() for x in ls1]
    for i,x in enumerate(brand):
        for i,e in enumerate(dataset["SKU"].str.lower()):
            try:
                if x in e:
                    dataset.loc[i,["Brand"]]=x
            except:
                print()
    main_data.append(dataset)

df=pd.concat(main_data)
try:
    df["date"]=df["date"].str.split(" ")
    df["date"]=[x[0] for x in df["date"]]
except:
    print()
brandrank=df.groupby(['date','Brand'])['Rank'].mean().reset_index()
skurank=df.groupby(['date','SKU'])['Rank'].mean().reset_index()
brandrank.columns=["date","Brand","brand rank"]
skurank.columns=["date","SKU","sku rank"]
df1=pd.merge(df,brandrank,on=["date","Brand"],how="inner")
final=pd.merge(df1,skurank,on=["date","SKU"],how="inner")
final.drop_duplicates(subset=["Brand","SKU","date"],inplace=True)
final.drop(["index","Rank",'sponsored_products'],axis=1,inplace=True)
for x in final["Brand"].unique():
    if x not in brand:
        print(x)
        final.drop((final["Brand"]==x).index)
final.to_csv("output.csv",index=False)
    
```
